```python
import time
from snake_env import SnakeEnv
import random
```

```python
def play_snake():
    env = SnakeEnv(grid_size=5, render_mode="human")
    obs, info = env.reset()
    print("Le jeu choisit des actions aléatoires.")

    for _ in range(10):
        action = random.choice([0, 1, 2, 3])  # Choisit une action aléatoire
        print(f"Action choisie: {action}")

        obs, reward, terminated, truncated, info = env.step(action)
        env.render()
        print(f"Récompense: {reward}")

        if terminated or truncated:
            print("Partie terminée!")
            break

        time.sleep(0.5)

play_snake()
```
